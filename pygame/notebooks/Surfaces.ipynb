```python
# here be no dragons
import pygame
import inspect
from pygame.locals import *

IMG_PATH = "img/matterhorn.jpg"
surface = pygame.image.load(IMG_PATH)
```

```python
pixel_array = pygame.PixelArray(surface)

for row in pixel_array:
    for i in range(0, len(row)):
        row[i] = 0
    break

for row in pixel_array:
    print(row)
    
# looks like PixelArray has following structure: 
# main PixelArray(row0 PixelArray, row1 PixelArray, row2 PixelArray, etc)
```
